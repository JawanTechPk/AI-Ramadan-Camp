```python
cities = ["Karachi","Lahore","Sukkur","Quetta","DIKhan"]
cityTocheck = input("Enter your city name: ")
if cityTocheck == cities[1]:
    print("It is clean city")
elif cityTocheck == cities[2]:
    print("It is clean city")
elif cityTocheck == cities[3]:
    print("It is clean city")
elif cityTocheck == cities[4]:
    print("It is clean city")
else:
    print("It is not a clean city")
```

```python
cities = ["Karachi","Lahore","Sukkur","Quetta","DIKhan"]

cleanCity = input("Enter your city name: ")

for cityToCheck in cities:
    if cleanCity == cityToCheck:
        print("It is a clean city")
```

**Nested For Loop**

```python
firstName = ["Fariha","Dure","Uroosa","Nida"]
lastName = ["Samara","Tooba","Aqsa","Sana"]
fullName = []
for first in firstName:
    for last in lastName:
        fullName.append(first + " " + last)
print(fullName)
```

**Dictionary inside list**

```python
quiz = [
    {
        "Sno" : "01",
        "question" : "What is your country name",
        "options" : ["Pakistan","China","Singapore","Iran"],
        "answer" : "Pakistan",
    },
    {
        "Sno" : "02",
        "question" : "What is your  name",
        "options" : ["Ahmed","Ali","Amjad","Adnan"],
        "answer" : "Ali",
    },
    {
        "Sno" : "03",
        "question" : "What is your age",
        "options" : ["22","18","28","32"],
        "answer" : "18",
    }
]

new_question = {
        "Sno" : "04",
        "question" : "What is your qualification",
        "options" : ["Masters","Graduate","Mphil","Phd"],
        "answer" : "Mphil"
}

quiz.append(new_question)


score = 0

for newQuiz in quiz:
    data = newQuiz["Sno"] + " : " + newQuiz["question"] + "\n" + newQuiz["options"][0]+ "\n" + newQuiz["options"][1] + "\n" + newQuiz["options"][2] + "\n" + newQuiz["options"][3] + "\n"
    ans = input(data)
    if ans == newQuiz["answer"]:
        score+=1
print("Your score for this quiz is : ",score)
```
